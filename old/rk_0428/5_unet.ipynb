```python
from Unet import unet
import numpy as np
from os.path import join
from os import walk
from all_p import ClusterFile
```

```python
model = unet(input_size=(2048, 2048, 5), data_format='channels_last')
```

```python
def data_gen(files, dirf, j=None):
    i = 0
    if not (j is None):
        i = j
    lenf = len(files)
    while True:
        
        datam = np.load(join(dirf, files[i]))
        
        data = np.array([datam[:,:,:-1]])
        mask = np.array([datam[:,:,-1]]).astype(np.uint8)
        
        i += 1
        i %= lenf
        
        if not (j is None):
            i = j
        yield data, mask
```

```python
def check_gen(gen, i=None):
    import matplotlib.pyplot as plt
    from skimage.io import imshow
    from show_pic import show_pic
    fig = plt.figure(figsize=(40, 20))
    data, mask = next(gen, i)
    fig.add_subplot(1, 3, 1)
    print(data.shape)
    print(data.min(), data.max())
    imshow(show_pic(data[0], idx=[0, 1, 2], sigma=0.5).astype(np.int32))
    fig.add_subplot(1, 3, 2)
    imshow(show_pic(data[0], idx=[2, 3, 4], sigma=0.5).astype(np.int32))
    print(mask.min(), mask.max())

    fig.add_subplot(1, 3, 3)
    imshow(mask[0,:,:])
```

```python
dirf = '/home/rt2122/data/train/filters/'
files = next(walk(dirf))[-1]
files = list(filter(lambda name:ClusterFile(name).params['id_list']==1, files))
files
```

```python
check_gen(data_gen(files, dirf))
```

```python
model.fit_generator(data_gen(files, dirf), steps_per_epoch=len(files), epochs=5)
```
