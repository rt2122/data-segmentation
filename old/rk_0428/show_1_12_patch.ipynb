```python
import numpy as np
```

```python
patch = np.load('/home/rt2122/data/train/filters/tf_il1_ip16_ra155.6608_dec48.9188_sgrizym_in5.npy')
```

```python
import matplotlib.pyplot as plt
```

```python
from skimage.io import imshow
```

```python
def show_mtx(mtx, idx=[0,1,-1]):
    pic = np.copy(mtx[:,:,idx])
    for i in range(3):
        pic[:,:,i] -= pic[:,:,i].min()
        pic[:,:,i] /= pic[:,:,i].max()
    imshow(pic)
```

```python
plt.figure(figsize=(8,8))
show_mtx(patch)
```
